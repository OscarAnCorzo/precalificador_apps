<h2>Modificación del Conjunto de Datos</h2>

<h3>- Renombrar algunas columnas </h3>

Para facilitar la implementación del conjunto de datos se arreglaron los nombres y se tradujeron al español. 

```python
d = d.rename(columns={'App':'nombre','Category': 'categoria','Rating':'calificacion','Reviews':'comentarios','Size':'tamaño',
                      'Installs':'instalaciones','Type':'tipo','Price':'precio','Content Rating':'clasificacion',
                      'Genres':'genero','Last Updated':'ult_act','Current Ver':'v_actual','Android Ver':'v_android'})
```

```python
d.columns
```

<h3>- Eliminación de caracteristicas. </h3>

Siguiendo el objetivo del modelo, son innecesarias las siguientes columnas:
- instalaciones.
- ult_act.
- v_actual.

```python
del d['instalaciones']
del d['ult_act']
del d['v_actual']
del d['comentarios']
```

```python
col = list(d.columns.values)
col
```

```python
d = d[['nombre',
 'categoria',
 'tama\xc3\xb1o',
 'tipo',
 'precio',
 'clasificacion',
 'genero',
 'v_android',
 'calificacion']]
```

```python
d
```
